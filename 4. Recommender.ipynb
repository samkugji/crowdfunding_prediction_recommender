```python
description_all = pd.DataFrame()
for i in tb_df['id']:
    try:
        response = requests.get('https://www.tumblbug.com/project/{number}'.format(number=i))
        dom = BeautifulSoup(response.text, 'html.parser')
        funding_type = pd.DataFrame([dom.select('ul.i-presents h3.b-panel__title')])
        des_df = pd.DataFrame([dom.select('div.b-content-description')[0].text], columns=['description'])
        des_type = pd.concat([des_df, funding_type], axis=1)
        des_type['id'] = i
        description_all = description_all.append(des_type)
        print(i)
    except:
        continue
```

```python
description_all.index = np.arange(len(description_all))
```

```python
description_all.to_csv('description_all.csv')
```
